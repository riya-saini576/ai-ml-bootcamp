# Part 2 – Handling Missing Values

---

## What are Missing Values?

A missing value is a value that is not available in the dataset.

Pandas represents missing values using **NaN (Not a Number)**.

Example

| Name | Age |
|------|-----|
| Alice | 25 |
| Bob | NaN |

Here, Bob's age is missing.

---

## Why are Missing Values Dangerous?

Missing values can:

- Produce incorrect statistics
- Cause machine learning models to fail
- Affect visualizations
- Reduce data quality

Therefore, they should always be identified before analysis.

```python
employee
```

```python
employee.isnull()
```

### Explanation

`isnull()` checks every cell.

If the value is missing, it returns **True**.

Otherwise, it returns **False**.

```python
employee.isna()
```

### Note

`isnull()` and `isna()` perform the same task.

Both return identical results.

```python
employee.notnull()
```

```python
employee.notna()
```

```python
employee.isnull().sum()
```

## Explanation

This is one of the most frequently used commands in Pandas.

`isnull()` identifies missing values.

`sum()` counts them.

The output shows how many missing values exist in each column.

```python
employee.isnull().sum().sort_values(ascending=False)
```

```python
employee.isnull().sum().sum()
```

## Total Missing Values

The first `sum()` counts missing values in each column.

The second `sum()` adds them together.

This gives the total number of missing values in the entire dataset.

```python
(employee.isnull().sum() / len(employee)) * 100
```

```python
round((employee.isnull().sum() / len(employee)) * 100, 2)
```

## Missing Percentage

Instead of counting missing values, we often calculate the percentage.

Formula

Missing Percentage

=

(Number of Missing Values / Total Rows) × 100

```python
missing_percentage = round(
    (employee.isnull().sum() / len(employee)) * 100,
    2
)

missing_percentage
```

```python
missing_percentage.plot(
    kind="bar",
    figsize=(10,5),
    title="Missing Values (%)"
)

plt.ylabel("Percentage")
plt.show()
```

## Visualization

The taller the bar,

the more missing values that column contains.

Visualizations help identify problematic columns quickly.

```python
employee["Age"].isnull()
```

```python
employee["Age"].isnull().sum()
```

```python
employee["Name"].isnull().sum()
```

## Mean vs Median vs Mode

Mean

Average value

Median

Middle value

Mode

Most frequently occurring value

```python
employee_mean = employee.copy()

employee_mean["Age"] = employee_mean["Age"].fillna(
    employee_mean["Age"].mean()
)

employee_mean
```

```python
employee_median = employee.copy()

employee_median["Age"] = employee_median["Age"].fillna(
    employee_median["Age"].median()
)

employee_median
```

```python
employee_mode = employee.copy()

employee_mode["Age"] = 0 if employee_mode["Age"] < 0 else pass

employee_mode
```

## Which Method Should We Use?

Numeric Columns

Mean or Median

Categorical Columns

Mode

Text Columns

Unknown or Not Available

There is no single best method.

It depends on the dataset.

```python
employee.fillna(0)
```

```python
employee.fillna("Not Available")
```

```python
employee.fillna({
    "Name":"Unknown",
    "Age":0,
    "Joining_Date":"Not Available"
})
```

# Part 3 - Handling Duplicate Records

---

## What are Duplicate Records?

Duplicate records are rows that appear more than once in a dataset.

Duplicates may occur due to

- Manual data entry
- Importing the same file multiple times
- System errors
- Database synchronization issues

Duplicates can affect

- Reports
- Data Analysis
- Machine Learning Models

Therefore, they should be identified and handled carefully.

```python
employee
```

## Observe the Dataset

Look carefully.

Employee ID **104** appears twice.

Both records are identical.

This is called a duplicate record.

```python
employee.duplicated()
```

## Explanation

The `duplicated()` function checks every row.

- False → Row appears for the first time.
- True → Row is a duplicate of a previous row.

By default, Pandas considers the **first occurrence as original**.

```python
employee[employee.duplicated()]
```

## Explanation

This displays only the duplicate rows.

```python
employee.duplicated().sum()
```

## Explanation

Count the total number of duplicate rows.

```python
print("Total Duplicate Rows :", employee.duplicated().sum())
```

```python
employee.drop_duplicates()
```

## drop_duplicates()

This function removes duplicate rows.

By default

- Keeps the first occurrence.
- Removes the remaining duplicates.

```python
employee
```

Notice that the original DataFrame is unchanged because we did not use

- inplace=True

or assign it back to a variable.

```python
employee_clean = employee.drop_duplicates()

employee_clean
```

```python
employee_clean.shape
```

```python
employee.shape
```

## Explanation

Original Rows

5

After Removing Duplicate

4

The **keep** parameter decides which duplicate row should be retained.

Options

- keep='first'
- keep='last'
- keep=False

```python
employee.duplicated(keep="first")
```

Default Behavior

The first occurrence is kept.

Remaining duplicate rows become True.

```python
employee[employee.duplicated(keep="first")]
```

```python
employee.duplicated(keep="last")
```

Now the last occurrence is considered original.

Earlier duplicate rows become True.

```python
employee[employee.duplicated(keep="last")]
```

```python
employee.duplicated(keep=False)
```

keep=False

Marks **every duplicate row** as True.

Neither first nor last is treated as original.

```python
employee[employee.duplicated(keep=False)]
```

```python
employee.drop_duplicates(keep="first")
```

```python
employee.drop_duplicates(keep="last")
```

```python
employee.drop_duplicates(keep=False)
```

Difference

keep="first"

Keeps first duplicate

------------------------

keep="last"

Keeps last duplicate

------------------------

keep=False

Removes all duplicate rows

Sometimes the entire row is not duplicated.

```python
labels = ["Unique", "Duplicate"]

values = [
    len(employee) - employee.duplicated().sum(),
    employee.duplicated().sum()
]

plt.figure(figsize=(5,4))

plt.bar(labels, values)

plt.title("Duplicate Records")

plt.ylabel("Count")

plt.show()
```
