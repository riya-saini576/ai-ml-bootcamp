```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
```

```python
np.random.seed(42)
n= 10000

df = pd.DataFrame({
    "Age" : np.random.randint(17,26,n),
    "Gender": np.random.choice(["Male","Female"],n),
    "Attendance":np.random.randint(40,101,n),
    "CGPA":np.round(np.random.uniform(4.5,10,n),2),
    "Backlogs":np.random.randint(0,8,n),
    "FamilyIncome":np.random.randint(150000,1200000,n),
    "StudyHour":np.round(np.random.uniform(0.5,8,n),1),
    "Scholarship":np.random.choice([0,1],n,p=[0.7,0.3]),
    "InternetAccess":np.random.choice([0,1],n,p=[0.15,0.85]),
    "Hostel":np.random.choice([0,1],n),
    "MentalStress":np.random.randint(1,11,n),
    "Partimejob":np.random.choice([0,1],n,p=[0.75,0.25])
}
)
    
   
    
    
    
```

```python
#Low attendance
risk = (
    (df["Attendance"] <65).astype(int)+
    (df["CGPA"] <6.0).astype(int)+
    (df["Backlogs"] >=3).astype(int)+
    (df["MentalStress"] >=8).astype(int)+
    (df["StudyHour"] <2).astype(int)
)
df["Dropout"] = np.where(risk>=3,1,0)
    
```

```python
df
```

```python
X =df.drop("Dropout",axis=1)
y=df["Dropout"]
```

```python
X
```

```python
X = pd.get_dummies(X,drop_first=True)
```

```python
X
```

```python
X_train,X_test,y_train,y_test = train_test_split(
    X,
    y,
    test_size = 0.2,
    random_state = 41,
    stratify =y
)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from xgboost import XGBClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
models = {
    "Logistic Regression": LogisticRegression(max_iter=1000),
    "KNN":KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "SVM": SVC(),
    "XGBoost": XGBClassifier(),
    "Naive Bayes": GaussianNB()  
}
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

results = []

for name, model in models.items():

    # Train
    model.fit(X_train, y_train)

    # Predict class labels
    y_pred = model.predict(X_test)

    # Predict probabilities
    if hasattr(model, "predict_proba"):
        y_prob = model.predict_proba(X_test)[:, 1]
    else:
        # SVM without probability=True
        y_prob = y_pred

    results.append({
        "Model": name,
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1 Score": f1_score(y_test, y_pred),
        "ROC AUC": roc_auc_score(y_test, y_prob)
    })
```

```python
result_df = pd.DataFrame(results)
result_df = result_df.sort_values(
    by="Accuracy",
    ascending = False
)
result_df
```
