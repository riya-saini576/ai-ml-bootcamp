```python
import numpy as np 
import pandas as pd 
np.random.seed(45)
n_student = 10000
data = {
    "Name" : np.random.choice(["dev","riya","ram"]),
    "Age" : np .random.randint(15,30,n_student),
    "city" :  np.random.choice(["ynr","amabala","kuk"]),
}
df= pd.DataFrame(data)
```

```python
import sqlite3 as sql
import pandas as pd
```

```python
# Connect to a database[create if it doesnot exist]
connection =sql.connect ('demo3.db')
```

```python
cursor =  connection.cursor()
```

```python
cursor.execute ('''CREATE TABLE IF NOT EXISTS PEOPLE(
    id INTEGER PRIMARY KEY,
    name TEXT NOT NULL,
    age INTEGER,
    city TEXT
);''')
```

```python
connection.commit()
```

```python
student_data = list(df.itertuples(index=False, name = None))
cursor.executemany('''INSERT INTO people (name,age,city) VALUES (?,?,?)''',student_data)


#cursor.execute('''INSERT INTO people (name,age,city) VALUES("DEV",32,"'YNR");''')
```

```python
connection.commit()
```

```python
cursor.execute('''SELECT*FROM people''')
```

```python
results= cursor.fetchall()
```

```python
for record in results:
    print(record)
```

```python
df = pd.DataFrame(results)
```
